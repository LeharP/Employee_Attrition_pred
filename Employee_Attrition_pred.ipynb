## Splitting the dataset

```python
from sklearn.model_selection import train_test_split
```

```python
X = data.drop(columns = ['Attrition_Yes'])
```

```python
X.head()
```

```python
y = data['Attrition_Yes']
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, random_state = 0)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale = StandardScaler()
X_train = scale.fit_transform(X_train)
X_test = scale.transform(X_test)
```

```python
X_train, X_test
```

# Model Selection

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
logi = LogisticRegression()
logi.fit(X_train,y_train)
y_pred = logi.predict(X_test)
```

```python
accu= accuracy_score(y_test,y_pred)
accu
```

```python
confusion_matrix(y_test,y_pred)
```

### Random Forest Classification

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
forest = RandomForestClassifier()
forest.fit(X_train,y_train)
y_pred = forest.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

### Support Vector Machine Classification

```python
from sklearn.svm import SVC
```

```python
xgb = SVC()
xgb.fit(X_train,y_train)
y_pred = xgb.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

# Logistic regression is the Best model

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
parameters = {
    'penalty':[ 'l2', None],
    'C':[0.25,0.5,0.75,1,1.25,1.5,1.75,2],
    'solver':['lbfgs', 'newton-cg', 'newton-cholesky', 'sag'],
    'max_iter':[50,100,1000,2000],
    'multi_class':['auto', 'ovr', 'multinomial']
}
```

```python
rand = RandomizedSearchCV(estimator= logi,param_distributions=parameters, n_iter=10, n_jobs=-1,cv =5, scoring = 'roc_auc',verbose=3, random_state=0)
```

```python
rand.fit(X_train,y_train)
```

```python
rand.best_estimator_
```

```python
rand.best_params_
```

```python
rand.best_score_
```

# Final Model

```python
lo = LogisticRegression(C=0.5, multi_class='auto', solver='newton-cholesky')
lo.fit(X_train,y_train)
y_pred = lo.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```
